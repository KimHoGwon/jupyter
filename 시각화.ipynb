```python
import matplotlib.pyplot as plt

X=range(100)
Y=range(100)
plt.plot(X,Y)
```

```python
import numpy as np

X_1 = range(100)
Y_1 = [np.cos(value) for value in X]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show()
```

```python
fig,ax = plt.subplots()
X1=range(100)
Y1=[np.cos(value) for value in X]
ax.plot(X1,Y1)
ax.set(title='cos graph',
       xlabel='X',
       ylabel='Y')
plt.show()
```

```python
fig=plt.figure()
fig.set_size_inches(10,10)
ax1=fig.add_subplot(1,2,1)
ax2=fig.add_subplot(1,2,2)

ax1.plot(X1,Y1,c="b")
ax2.plot(X_2,Y_2,c="g")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(-1,1,100)
y1=np.sin(x)
y2=np.cos(x)
y3=np.tan(x)
y4=np.exp(x)

fig, ax = plt.subplots(2,2)

ax[0,0].plot(x,y1)
ax[0,1].plot(x,y2)
ax[1,0].plot(x,y3)
ax[1,1].plot(x,y4)

plt.show()
```

```python
ax1=plt.subplot(321)
ax1.plot(x,y1)
ax2=plt.subplot(322)
ax2.plot(x,y1)
ax3=plt.subplot(312)
ax3.plot(x,y1)
ax4=plt.subplot(325)
ax4.plot(x,y1)
ax5=plt.subplot(326)
ax5.plot(x,y1)

plt.show()
```

```python
x1=range(100)
y1=[value for value in x1]

x2=range(100)
y2=[value+100 for value in x2]

plt.plot(x1, y1, color="#000000")
plt.plot(x2, y2, c="c")

plt.show()
```

```python
plt.plot(x1, y1, c="b", linestyle="dashed")
plt.plot(x2, y2, c="r", ls="dotted")
plt.title("Two lines")
plt.show()
```

```python
flg = plt.figure()
flg.set_size_inches(10, 10)

ax_1 = flg.add_subplot(1, 2, 1)
ax_2 = flg.add_subplot(1, 2, 2)

ax_1.plot(x1, y1, c="b")
ax_1.set_title("Figure 1")
ax_2.plot(x2, y2, c="g")
ax_2.set_title("Figure 2")

plt.show()
```

```python
plt.plot(x1, y1, color="b", linestyle="dashed", label='line_1')
plt.plot(x2, y2, color="r", linestyle="dotted", label='line_2')
plt.legend(shadow=True, fancybox=False, loc="upper right")

plt.title('$y = ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')
```

```python
data=[[5.,25.,50.,20.],
      [4., 34., 52., 27],
      [6., 22., 52., 19]]

x=np.arange(0,8,2)

plt.bar(x+0.00, data[0], color='b', width=0.50)
plt.bar(x+0.50, data[1], color='g', width=0.50)
plt.bar(x+1.0, data[2], color='r', width=0.50)

plt.xticks(x+0.50, ("A","B","C","D"))

plt.show()
```

```python
data = np.array([[5., 5., 5., 5.],
                [4., 34., 51., 27],
                [6., 22., 52., 19]])

color_list=['b', 'g', 'r']
data_label=["A","B","C"]

x= np.arange(data.shape[1])

for i in range(3):
    plt.bar(x, data[i],
            bottom = np.sum(
                data[:i], axis=0),
            color=color_list[i],
            label=data_label[i])
plt.legend()
plt.show()
```
